```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Function to extract each product title
def get_title(new_soup):
    
    try:
        
        #Outer Tag Object
        title = new_soup.find("span", attrs ={"class":"a-size-large product-title-word-break"})
        
        #Inner Navigatable String object
        title_text = title.text
        
        #Title as a String Value
        title_string = title_text.strip()
        
    except AttributeError:
        title_string = ""
        
    return title_string

#Function to extract each Product Price
def get_price(new_soup):
    try:
        
        price = new_soup.find('span', attrs = {'class' : 'a-offscreen'})
        price_text = price.text
        price_string = price_text.strip()
        
    except AttributeError:
        price_string = ""
    
    return price_string

#Function to extract each Rating

def get_rating(new_soup):
    try:
        
        rating = new_soup.find('span',attrs = {'class':'a-icon-alt'})
        rating_text = rating.text
        rating_strp = rating_text.strip()
        
    except AttributeError:
        rating_strp = ""
    
    return rating_strp

def get_rating_count(new_soup):
    
    try:
        ratings = new_soup.find('span', attrs ={'id': 'acrCustomerReviewText'})
        ratings_text = ratings.text
        
    except AttributeError:
        ratings_text = ""
        
    return ratings_text

def get_color(new_soup):
    
    try:
        color = new_soup.find("span", attrs = {'class':'selection'}).text.strip()
    except:
        color = ""
    return color
```

```python
if __name__ == "__main__":
    
    #Headers for request
    HEADERS = ({"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36 Edg/111.0.1661.54",'Accept-Language' : 'en-US, en; q = 0.5'})
    
    #Webpage URL
    URL = "https://www.amazon.com/s?k=mobiles+phones&crid=KYXL4I8ELKTN&sprefix=mobiles+phone%2Caps%2C318&ref=nb_sb_noss_2"
    
    #HTTP Request
    webpage = requests.get(URL, headers = HEADERS)
    
    
    #Soup Object containing all data
    soup = BeautifulSoup(webpage.content, 'html.parser')
    
    #Fetch Links as list of tag objects
    links = soup.find_all("a", attrs = {'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
    
    #Store the links
    links_list = []
    
    #Loop for extracting links from tag objects
    for link in links:
        links_list.append(link.get('href'))

    dict_info = {'title': [], 'price' : [], 'color':[], 'rating': [], 'rating_count' : []}
    
    #Loop for extracting product details for each link
    for link in links_list:
        new_webpage = requests.get("https://amazon.com"+link, headers = HEADERS)
        new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
        
        
        #Function Calls to display all necessary info
        dict_info['title'].append(get_title(new_soup))
        dict_info['price'].append(get_price(new_soup))
        dict_info['color'].append(get_color(new_soup))
        dict_info['rating'].append(get_rating(new_soup))
        dict_info['rating_count'].append(get_rating_count(new_soup))
    
    amazon_df = pd.DataFrame.from_dict(dict_info)
    amazon_df.to_csv("amazon_data.csv", header = True, index = False)
```
